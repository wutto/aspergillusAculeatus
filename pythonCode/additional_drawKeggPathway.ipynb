```python
##create function to convert raw data to percent
def hextriplet(colortuple):
    return '#' + ''.join(f'{i:02X}' for i in colortuple)

# A bit of code that will help us display the PDF output
def PDF(filename):
    return HTML('<iframe src=%s width=700 height=350></iframe>' % filename)

# A bit of helper code to shorten long text
def head(text, lines=10):
    """ Print the first lines lines of the passed text.
    """
    print ('\n'.join(text.split('\n')[:lines] + ['[...]']))
```

```python
##Load output file to create report table
path = "./"##Please define directory that stored your input file
os.chdir(path)

inputFile = "user_ko.txt"
b = open(inputFile,'r').readlines()
ko = {}
for i in b:
    if "\t" in i:
        key = i.replace("\n","").split("\t")[-1]
        ##value = kegg_list(key).read().split("\t")[-1].replace("\n","")
        value = ""
        ko[key] = value
        #print (key,'\t',value)
        
##Load list of KEGG pathway
b = open("listKeggPathway.txt",'r').readlines()
del b[0]##delete header

pathways = {}
for i in b:
    iList = i.replace("\n","").split("\t")
    key = iList[0]
    value = [iList[1],iList[2]]##value[0] ==> description,value[1] ==> group
    pathways[key] = value
    
```

```python
print ("NO. KO:",len(ko))
print ("NO. pathway:",len(pathways))
```

```python
##Create folder to keep pathway image
pathPathway = path+"/pathwayImage/"
if os.path.exists(pathPathway) == False:
    os.mkdir(pathPathway)
os.chdir(pathPathway)
```

```python
##Draw pathway and label color
for pathwayId in pathways.keys():
    
    """create output fileName"""
    saveName = pathways[pathwayId][1].replace(" ","_")+"_"+pathwayId+".pdf"
    print (saveName)
    
    """Loading pathway information"""
    pathway = KGML_parser.read(kegg_get(pathwayId, "kgml"))
    canvas = KGMLCanvas(pathway, import_imagemap=True)

    """change the colours of ortholog elements"""
    # Change the colours of ortholog elements
    for element in pathway.orthologs:
        koId = element.name.split(":")[-1]
        if koId in ko:
            log2 = ko[koId]
            for graphic in element.graphics:
                graphic.bgcolor = "#00BFFF"
        else:
            for graphic in element.graphics:
                graphic.bgcolor = "#FFFFFF"

    canvas.draw(saveName)
    
```

```python
##Merge all pathways into single file
from PyPDF2 import PdfFileMerger

pdfs = os.listdir("./")

merger = PdfFileMerger()

for pdf in pdfs:
    merger.append(pdf)
    print (pdf)

merger.write("../keggPathways.pdf")
merger.close()
```
